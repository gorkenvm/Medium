# Veri Bilimi ve CRİSP DM Metodolojisi


CRoss Industry Standard Process for Data Mining (CRISP-DM)

Veri bilimi projesi yapmaya başladığım ve süreçler içerisinde kaybolduğum anda tanıştığım ve hayatımı kurtaran, temel görevi gören bir süreç modelidir.

Projeye başlarken adım adım neler yapılması gerektiği ve proje içerisinde beklenmeyen sonuçlar alındığı zaman hangi sürece geri dönmen gerektiğini dikkatle incelemeni tavsiye ederim.

CRISP-DM metodolojisinin Altı ardışık aşaması vardır.

## 1. Business Understanding

Projenin amaçlarının ve gereksinimlerinin ne olduğunu anlama bölümüdür.

- Müşterinin gerçekten neyi başarmak istediğini anla, Hedefini belirle
- Gereksinimlerini belirle, Nasıl bir veriye ihtiyacın var ? Kısıtlamalar nelerdir ?
- İş sonunda nasıl bir çıktı hedefliyorsun ? Projenin başarılı olması nedir ?
- Hangi teknolojilere ihtiyacın var ?

## 2. Data Understanding

Bu bölümde belirlenen verileri topla ve analiz et.

- İlk olarak bağımlı ve bağımsız değişkenleri belirle.
- Değişkenleri Sayısal, Kategorik ve Kardinal olarak grupla. Bu gruplama diğer bölümlerde işlerini kolaylaştıracak.
- Genel bir bakış ile verinin istatistiksel (mean, std, quantile ..) dağılımını incele.
- Her bir bağımsız değişkenin, bağımlı değişken ile ilişkisini incele. Görselleştirmeler de kullanmanı tavsiye ederim.
- İki yada daha fazla bağımsız değişkenin bağımlı değişkenle ilişkisini incele. Görselleştirmeler de kullanmanı tavsiye ederim.
- Veri kalite kontrolü, yani veri kirli ve gürültülü ise nasıl temizleyebileceğini tanımla

Bu kısımda sorun yaşarsan bir önceki bölüm olan business understanding’e geri dönmen gerekebilir.

Not : Bağımlı değişken, tahmin edeceğin değişken. Bağımsız değişkenler, bağımlı değişkeni bulmak için modelin örüntü yakalaması gereken değişkenlerdir.

Not : Gruplama yaparken ‘’Sinsirella’’ diye nitelendirdiğim kategorik görünümlü ancak sayısal, sayısal görünümlü ancak kategorik değişkenlere dikkat et.

## 3. Data Preparation


Projede en çok vakit harcayacağınız, dönüp dönüp tekrar geleceğiniz kısım denilebilir.

- Değişken seçimi : hangi değişkenlerin kalacağı ve çıkarılacağını belirle
- Eksik verileri tespit ederek gerekli yöntemlerle doldur, doldurma işlemi mümkün görünmüyorsa ve verinin %1–2 sini oluşturuyorsa veriden çıkarabilirsin. Ancak unutma veri silmek en son tercihimiz olmalıdır.
- Aykırı gözlemlerin tespit edilmesi ve baskılanması. Ortalaması 50 maksimum değeri 100 olan bir değişkende 200 gibi bir değer olabilir ancak gerçek hayatta bunun gözlemlenme frekansı düşük olacaktır. Amacımız genel bir örüntü yakalamak olduğu için bu değerleri alt veya üst sınırlarla baskılayabilirsin.
- Feature engineering : Değişken mühendisliği, elimizdeki değişkenleri kullanarak farklı değişkenler oluşturma sanatıdır. Bu kısımda iş bilgisi oldukça önem kazanıyor.
- Encoding, makine öğrenmesi modelleri sayısal değişkenlerle çalışabilmektedir. Kategorik değişkenleri farklı birer değişken haline getirmek için yaptığımız işlemdir.
- Scaling, verideki anlamın değişmemesi şartı ile tüm değişken değerlinin aynı formata getirilmesidir. Bu sayede modelimiz daha hızlı çalışacak ve oluşabilecek yanlılıklar ortadan kalkacak.

Bu kısımları iyi yaptıysanız hayırlı olsun proje bitti denilebilir.

Not : Buradaki her bir adım bir sonraki yazılarda uçtan uca bir proje ile yazıya alınacaktır.

## 4. Modeling


Model model dediğimiz heyecan veren kısma geldik.

- Kullanılacak Makine Öğrenmesi algoritmaları seçilecektir. Bulmaya çalıştığımız değişken sayısal ise Regresyon, sıfır birlerden oluşan kategorik bir değişken ise Sınıflandırma algoritmaları seçilir.
- Veriyi eğitim ve test olarak belirli oranlarda bölme işlemi yap. Eğitim verisi üzerinde model oluştur ve test verisi üzerinde modelin tahminlerini al. Bu tahmin sonuçlarını bir sonraki bölümde gerçek sonuçlar ile karşılaştıracağız.
- En iyi sonuçları yakalamak için mümkün oldukça fazla model kurulmasını tavsiye ederim.

Kurulan modellerde Aşırı öğrenme (Over-Fitting) veya Az öğrenme (Under-fitting) var ise veri iyi hazırlama bölümüne dönülmesi gerekecektir. Özellikle aşırı öğrenme durumunda, bağımlı değişken ile yüksek kolere bir değişken olabilir bunun veriden çıkarılmasını tavsiye ederim. Bu bölümde sorun yaşanması halinde bir önceki bölüm olan ver ön hazırlık bölümüne gidilmesi gerekecektir.

## 4. Evaluation
Doğruluk, Duyarlılık, Özgüllük, F-Skoru, AUC, R-Sq, Adj R-SQ, RMSE (Kök Ortalama Kare Hatası) gibi birçok model değerlendirme tekniği vardır ve liste devam etmektedir. Hangi değerlendirme metriklerinin tamamen seçileceği, değerlendirme kriterlerine, istenen nihai modele, iş gereksinimine ve kullanılan model algoritmasına bağlıdır.

Sonuçların iyileştirilmesi için veri ön hazırlama kısmı tekrardan incelenebilir yada model kurma bölümünde model tunning yapılmalıdır. Model tunning, algoritmaların hiper parametrelerinin değiştirilerek en idealinin bulunması işlemidir. Ayrıca Feature selection denilen değişken seçimi işleminin yapılması da gerekli olabilmektedir.

Bu bölümde sorun yaşanması durumunda ilk bölüm olan Business Understanding kısmına tekrar dönülmesi tavsiye edilir.

## 4. Deployment
Modelin başarısının yeterli kabul edildiği durumda uygulama kısmına geçilir. Ancak modelin kararlı olması modelin başarısından daha önemli olabilmektedir. Düşünün ki yüksek başarı ile tahmin yapan modelinizin çalışıp sonucu döndürmesi 5 dakika sürüyor. Böyle durumlarda hızı arttırmak için modelin başarısından feragat edilebilmektedir.
